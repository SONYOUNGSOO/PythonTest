#1.Open API
실시간 영업소간 통행시간
http://data.ex.co.kr/openapi/trtm/realUnitTrtm

```python
import requests
```

```python
key= '9433999045'
type = 'json'
StartUnitCode = '101'
EndUnitCode = '103'

URL = 'http://data.ex.co.kr/openapi/trtm/realUnitTrtm?key='
url = URL + key + '&type=json&iStartUnitCode='+StartUnitCode + '&iEndUnitCode' + EndUnitCode
```

```python
response = requests.get(url)
```

```python
print(response)
```

```python
json = response.json()
```

```python
json
```

```python
cars = json['realUnitTrtmVO']
```

```python
print(cars)
```

```python
cars
```

```python
records = []
for car in cars:
    dic = {}
    dic['data'] = car['stdDate']
    dic['time'] = car['stdTime']
    dic['type'] = car['tcsCarTypeDivName']
    records.append(dic)
records
```
